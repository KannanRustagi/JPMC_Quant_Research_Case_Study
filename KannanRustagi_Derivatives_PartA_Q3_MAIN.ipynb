```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#code for plotting the variation of Payoff of Trader X along with stock price in Part-A Q3
x=[1000, 600, 1400]
y=[0, 400, -400]
plt.plot(x, y)
plt.title('Variation of Payoff of Trader X along with stock price')
plt.xlabel('Stock Price($)', color='#1C2833')
plt.ylabel('Payoff($)', color='#1C2833')
plt.plot(1000, 0, marker="o", markersize=7, markeredgecolor="black", markerfacecolor="red")
plt.annotate("Futures Price", (1000,0))
plt.grid()
plt.show()
```

```python
#code for plotting the variation of Payoff of Trader Y along with stock price in Part-A Q3
x=[500, 1000, 1100, 1500]
y=[400, -100, -100, -100]
plt.plot(x,y)
plt.title('Variation of Payoff of Trader Y along with stock price')
plt.xlabel('Stock Price($)', color='#1C2833')
plt.ylabel('Payoff($)', color='#1C2833')
plt.plot(1000, -100, marker="o", markersize=7, markeredgecolor="black", markerfacecolor="red")
plt.annotate("Strike Price", (1000,-100))
plt.grid()
plt.show()

```

```python
#code for plotting the comparison of payoffs of futures and options strategy
x1=[1000, 600, 1500]
y1=[0, 400, -500]
x2=[600, 1000, 1100, 1500]
y2=[300, -100, -100, -100]
plt.plot(x1, y1, color='r', label='futures payoff')
plt.plot(x2, y2, color='g', label='options payoff')
  
plt.xlabel("Stock Price($)")
plt.ylabel("Payoff($)")
plt.title("Comparison of futures and options payoff")
plt.axvspan(600, 1100, color='yellow', alpha=0.75, lw=0)
plt.axvspan(1100, 1500, color='pink', alpha=0.75, lw=0)
  

plt.legend()
plt.grid()
  
plt.show()
```
